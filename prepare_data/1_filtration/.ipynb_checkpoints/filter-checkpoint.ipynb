```python
plt.figure(figsize = (20, 10))
colors = ['firebrick', 'orangered', 'orange', 'pink', 'lime', 'green', 'aqua', 'blue', 'magenta', 'chocolate']

filter_dict = dict()


for index, sp in enumerate(sp_list):
    plt.subplot(2, 5, index + 1)
    
    temp = at_comp[at_comp['sp'] == sp]
    plt.hist(temp['at_comp'], bins = 100, density=True, color = colors[index], alpha = 0.3, label =sp )
    plt.hist(temp['at_comp'], bins = 100, density=True, color = colors[index], alpha = 1, histtype='step' )
    plt.title(sp)
    
    
    
    nparam_density = stats.kde.gaussian_kde(temp['at_comp'])
    x = np.linspace(0.2, 0.9, 5000)
    nparam_split = nparam_density(x)
    moda = x[np.argsort(nparam_split)[-1]]
    print(sp, moda, len(temp['at_comp']))
    
    right_vals = [elem for elem in list(temp['at_comp']) if elem >= moda] 
    


    right_std = np.std(right_vals)
    print(len(right_vals),right_std, moda - 4*right_std )
    plt.axvline(x = moda, color = 'black', linestyle = '--')
#     plt.axvline(x = moda - 2*right_std, color = 'grey', linestyle = '--')
#     plt.axvline(x = moda - 3*right_std, color = 'grey', linestyle = '--')
    
    if sp in ['raik', 'minu']:
        filter_dict[sp] = moda - 5*right_std
        plt.axvline(x = moda - 5*right_std, color = 'red', linestyle = '--')
    else:
        filter_dict[sp] = moda - 6*right_std
        plt.axvline(x = moda - 6*right_std, color = 'grey', linestyle = '--')
    
    
    plt.xlim(0.2, 0.9)

    

plt.show()
```

```python
def FilterRowBySpeciesAT(row, filter_dict):
    if row['at_comp'] > filter_dict[row['sp']]:
        return True
    else:
        return False
    
at_comp['fits'] = at_comp.apply(lambda row : FilterRowBySpeciesAT(row, filter_dict), axis = 1)
```

## Make Extended data table 

```python
raw = pd.read_csv('./blast_filter_table_creation/results/filter_blast_table.txt', header = 0, sep = ' ')
```

```python
raw['sp'] = raw.progress_apply(lambda row: row['idr'][:4], axis = 1)
```

```python
ext = pd.read_csv('ext.txt', header = 0, sep = '\t')
```

```python
ext
```

```python
def GetColumns(fancy_idr, fancy_acc):
    if type(fancy_acc) == float:
        fancy_acc = 'N'
    cols = list()
    for elem in raw.columns:
        if fancy_idr.lower() in elem.lower() or ("GCA" in fancy_acc and fancy_acc in elem) :
            cols.append(elem)
        elif ',' in fancy_acc:
            accs = fancy_acc.split(',')
            for acc in accs:
                if acc in elem:
                    cols.append(elem)
        elif 'Homo_sapiens' in fancy_idr:
            if 'human' in elem:
                cols.append(elem)
    return cols
            

            
eu_species = ['minu', 'foca', 'cras', 'octo', 'rari', 'raik', 'eury', 'petz', 'harp']



def GetNumber(sp, fancy_idr, fancy_acc):
    cols = GetColumns(fancy_idr, fancy_acc)
    temp = raw[raw['sp'] == sp]
    temp = temp[cols]
    return temp.sum().sum()/len(raw[raw['sp'] == sp]) * 100
```

```python
ls = list()
for sp in eu_species:
    temp = raw[raw['sp'] == sp]
    ls.append(str(len(temp)))

print('\t'.join(ls))
```

```python
for sp in eu_species:

    sp_list = list()
    for index, row in ext.iterrows():
        fancy_idr, fancy_acc = row['idr'], row['accession']
        sp_list.append(GetNumber(sp, fancy_idr, fancy_acc))
    
    ext[sp] = sp_list
```

```python
ext['sum'] = ext[eu_species].sum(axis = 1)
```
